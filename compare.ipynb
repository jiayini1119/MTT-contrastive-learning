```python
import matplotlib.pyplot as plt

def read_data(file_path):
    ipc_values = []
    grad_averages = []
    with open(file_path, "r") as f:
        for line in f:
            ipc, grad_avg = line.strip().split(", ")
            ipc_values.append(int(ipc.split(": ")[1]))
            grad_averages.append(abs(float(grad_avg.split(": ")[1])))
    return ipc_values, grad_averages

ipc_values_original, grad_averages_original = read_data("grad_averages_vs_ipc.txt")
ipc_values_crop, grad_averages_crop = read_data("grad_averages_vs_ipc_crop.txt")
real_ipc_values_crop, real_grad_averages_crop = read_data("real_grad_averages_vs_ipc_crop.txt")
real_ipc_values, real_grad_averages = read_data("real_grad_averages_vs_ipc.txt")




plt.plot(ipc_values_original, grad_averages_original, marker='o', label='Random Init with Crop')
plt.plot(ipc_values_crop, grad_averages_crop, marker='o', label='Random Init without Crop')
plt.plot(real_ipc_values_crop, real_grad_averages_crop, marker='o', label='Real Init without Crop')
plt.plot(real_ipc_values, real_grad_averages, marker='o', label='Real Init without Crop')

plt.xlabel('IPC')
plt.ylabel('Gradient Average')
plt.title('IPC vs Gradient Average')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

def read_data(file_path):
    N_values = []
    grad_averages = []
    with open(file_path, "r") as f:
        for line in f:
            N, grad_avg = line.strip().split(", ")
            N_values.append(int(N.split(": ")[1]))
            grad_averages.append(abs(float(grad_avg.split(": ")[1])))
    return N_values, grad_averages

N_values_original, grad_averages_original = read_data("grad_averages_vs_N.txt")
N_values_crop, grad_averages_crop = read_data("grad_averages_vs_N_crop.txt")
real_N_values_crop, real_grad_averages_crop = read_data("real_grad_averages_vs_N_crop.txt")
real_N_values, real_grad_averages = read_data("real_grad_averages_vs_N.txt")



plt.plot(N_values_original, grad_averages_original, marker='o', label='Random Init with Crop')
plt.plot(N_values_crop, grad_averages_crop, marker='o', label='Random Init without Crop')
plt.plot(real_N_values_crop, real_grad_averages_crop, marker='o', label='Real Init without Crop')
plt.plot(real_N_values, real_grad_averages, marker='o', label='Real Init with Crop')

plt.xlabel('N')
plt.ylabel('Gradient Average')
plt.title('Synthetic Steps vs Gradient Average')
plt.grid(True)
plt.legend()
plt.show()
```
