```python
import torch
import matplotlib.pyplot as plt

distilled_images_path = "/home/jennyni/MTT-contrastive-learning/distilled_data/distilled_images.pt"
labels_path = "/home/jennyni/MTT-contrastive-learning/distilled_data/labels.pt"
```

```python
distilled_images = torch.load(distilled_images_path)
labels = torch.load(labels_path)

num_classes = 100
ipc = 1

for i in range(num_classes):
    
    class_images = distilled_images[i * ipc: (i + 1) * ipc]
    class_labels = labels[i * ipc: (i + 1) * ipc]
        
    plt.figure(figsize=(20, 5))
    for j, img in enumerate(class_images):
        plt.subplot(1, ipc, j + 1)
        plt.imshow(img.detach().permute(1, 2, 0).numpy())
        plt.title(f"Class {class_labels[0].item()}")
        plt.axis('off')
    plt.show()
```
